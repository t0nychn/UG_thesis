```python
djci = load('data/saved/betas3.csv')
gsci = load('data/saved/betas.csv')
```

```python
#colours = ['tab:blue', 'tab:red', 'black']
xs = ['SPX', 'MSCI_W', 'MSCI_EM']
fig, ax = plt.subplots(3, sharex=True, figsize=(15,7))
for i in range(len(xs)):
    ax[i].set_title(f'{xs[i]} Beta')
    ax[i].plot(djci[f'DJCI-{xs[i]}-smth'], color='tab:purple', label='DJCI')
    ax[i].plot(gsci[f'GSCI-{xs[i]}-smth'], color='tab:blue', label='GSCI')
    ax[i].axhline(0, color='grey', linestyle='--')
    ax[i].legend()
```

```python
djci_p = load('data/saved/probs3.csv')
gsci_p = load('data/saved/probs.csv')
xs = ['SPX', 'MSCI_W', 'MSCI_EM']
fig, ax = plt.subplots(3, sharex=True, figsize=(15,7))
for i in range(len(xs)):
    ax[i].set_title(f'P({xs[i]} Beta) > 0')
    ax[i].plot(djci_p[f'DJCI-{xs[i]}-smth'], color='tab:purple', label='DJCI')
    ax[i].plot(gsci_p[f'GSCI-{xs[i]}-smth'], color='tab:blue', label='GSCI')
    ax[i].axhline(0, color='grey', linestyle='--')
    ax[i].legend()
```
