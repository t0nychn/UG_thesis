```python
betas = load('data/saved/betas.csv')
probs = load('data/saved/probs.csv')
```

```python
def sum_cols(df):
    cols = df.columns.to_list()
    sums = {}
    for col in cols:
        sums[col] = df[col].sum()
    return sums
```

# Betas

```python
beta_sums = sum_cols(betas)
```

```python
# rank
beta_champs = dict(sorted(beta_sums.items(), key=lambda item: item[1]))
plt.figure(figsize=(10, 30))
plt.barh(list(beta_champs.keys()), list(beta_champs.values()))
```

# Probs

```python
prob_sums = sum_cols(probs)
```

```python
# rank
prob_champs = dict(sorted(prob_sums.items(), key=lambda item: item[1]))
plt.figure(figsize=(10, 30))
plt.barh(list(prob_champs.keys()), list(prob_champs.values()))
```
