# Multicollinearity Check

```python
from utils import *
from statsmodels.graphics.tsaplots import plot_acf
```

```python
vix = load('data/VIX.csv')
vix_shock = calc_shock('VIX', vix)
```

```python
t1 = plot_acf(vix.VIX, lags=12, title='VIX autocorrelation')
t2 = plot_acf(vix_shock.VIX, lags=12, title='VIX shocks autocorrelation')
```

```python
epu = load('data/EPUGlobal.csv')
epu_shock = calc_shock('GEPU_current', epu)
t1 = plot_acf(epu.GEPU_current, lags=12, title='EPU autocorrelation')
t2 = plot_acf(epu_shock.GEPU_current, lags=12, title='EPU shocks autocorrelation')
```

```python
gpr = load('data/GPRGlobal.csv')
gpr_shock = calc_shock('GPRI', gpr)
t1 = plot_acf(gpr.GPRI, lags=12, title='GPR autocorrelation')
t2 = plot_acf(gpr_shock.GPRI, lags=12, title='GPR shocks autocorrelation')
```

Overall our regressors exhibit at most 0.25 autocorrelation when transformed into shocks. This is not high enough to warrant concerns of multicollinearity for our distributed lag model, hence we can use OLS to estimate response coefficients.
